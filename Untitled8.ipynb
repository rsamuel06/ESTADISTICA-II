```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
from google.colab import files # Para cargar archivos en google colab
uploaded = files.upload()
```

```python
#Cargando base de datos
datos_icfes = pd.read_csv(
    io.BytesIO(uploaded['SB11_20231.TXT']),
    delimiter='¬',
    engine='python'
)
```

```python
#1. Eliminación de valores faltantes
#Depurando base datos, eliminando valores faltantes
icfes_sin_nan = datos_icfes.dropna(subset=[
    'ESTU_INSE_INDIVIDUAL',
    'PUNT_LECTURA_CRITICA',
    'PUNT_MATEMATICAS',
    'PUNT_C_NATURALES',
    'PUNT_SOCIALES_CIUDADANAS',
    'PUNT_INGLES'
])
```

```python
#Mostrando los primeros 20 datos de la base de datos sin valores faltantes
icfes_sin_nan.head(20)
```

```python
#2. Selección de variables explicativas a partir de los diagramas de dispersión
# Definimos las variables explicativas como los puntajes en  cada areá y la variable respuesta como el índice socioeconómico
variables_explicativas = [
    'PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS',
    'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES'
]
variable_respuesta = 'ESTU_INSE_INDIVIDUAL'

# gráficos
plt.figure(figsize=(15, 10))

for i, var in enumerate(variables_explicativas, 1):
    plt.subplot(2, 3, i)
    # Usamos regplot para ver la linea de tendencia
    sns.regplot(x=var, y=variable_respuesta, data=icfes_sin_nan,
                scatter_kws={'alpha':0.2, 's':10}, line_kws={'color':'red'})
    plt.title(f'Relación: {var}')
    plt.xlabel('Puntaje')
    plt.ylabel('Índice Socioeconómico')

plt.tight_layout()
plt.show()
```

Se puede ver que las variables PUNT_INGLES y PUNT_MATEMATICAS presentan una tendencia ascendente más definida que las otras

```python
#3. Selección de variables explicativas a partir de los coeficientes de correlación
# Definimos las variables de estudio
variables_estudio = [
    'ESTU_INSE_INDIVIDUAL', 'PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS',
    'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES'
]

# Creamos un DataFrame solo con las columnas de interés
df_corr = icfes_sin_nan[variables_estudio]

# Calculamos los tres tipos de correlación
corr_pearson = df_corr.corr(method='pearson')[['ESTU_INSE_INDIVIDUAL']]
corr_kendall = df_corr.corr(method='kendall')[['ESTU_INSE_INDIVIDUAL']]
corr_spearman = df_corr.corr(method='spearman')[['ESTU_INSE_INDIVIDUAL']]

# Mostramos en una sola tabla para comparar
tabla_correlaciones = pd.concat([corr_pearson, corr_kendall, corr_spearman], axis=1)
tabla_correlaciones.columns = ['Pearson', 'Kendall', 'Spearman']

# Eliminamos la fila de la variable respuesta consigo misma (que siempre da 1) para que quede más claro
tabla_correlaciones = tabla_correlaciones.drop('ESTU_INSE_INDIVIDUAL')

print("Tabla Comparativa de Correlaciones:")
print(tabla_correlaciones.sort_values(by='Pearson', ascending=False))
```

PEARSON: Vemos que la variable PUNT_INGLES tiene una fuerza moderada de correlación ya que se encuentra entre 0.5 y 0.75 y las demas variables se quedan en una correlación débil.


KENDALL: Es la correlación con menos fuerza de las 3, pero la variable PUNT_INGLES con 0.39 tiene la probabilidad más alta de que, si tomas dos estudiantes al azar, el que tenga mayor puntaje en Inglés también tendrá el mayor índice socioeconómico.


SPEARMAN: El hecho de que Spearman sea mayor que Pearson confirma que hay valores atípicos en los puntajes que están afectando la linealidad, pero no la tendencia general y al ser el coeficiente mas alto en todas las categorias se puede ver que es una relación monótona

```python
#4. Selección de variables explicativas a partir del coeficiente de determinación calculados a partir del modelo de regresión lineal múltiple
# Variables explicativas
features_full = [
    'PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS',
    'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES'
]
target = 'ESTU_INSE_INDIVIDUAL'

# 1. R^2 total (con TODAS las variables)
X_all = icfes_sin_nan[features_full]
y = icfes_sin_nan[target]
modelo_full = LinearRegression().fit(X_all, y)
r2_total = modelo_full.score(X_all, y)

print(f"R^2 con todas las variables: {r2_total:.5f}\n")
print("Evaluación de importancia (omitiendo una a una):")
print("-" * 50)

# 2. R^2 reducido (omitiendo una variable cada vez)
resultados_r2 = []

for var_omitida in features_full:
    # Crear lista de variables excepto la actual
    features_reducidas = [v for v in features_full if v != var_omitida]

    X_reduced = icfes_sin_nan[features_reducidas]
    modelo_red = LinearRegression().fit(X_reduced, y)
    r2_reducido = modelo_red.score(X_reduced, y)

    # Calculamos la diferencia
    diferencia = r2_total - r2_reducido
    resultados_r2.append((var_omitida, r2_reducido, diferencia))

    print(f"Omitiendo {var_omitida:25} | R^2: {r2_reducido:.5f} | Disminución: {diferencia:.5f}")
```

```python
# Graficar diferencia
nombres, valores, _ = zip(*resultados_r2)
plt.figure(figsize=(10, 5))
plt.barh(nombres, [r2_total - v for v in valores], color='skyblue')
plt.title('Importancia de las variables (Pérdida de $R^2$ al omitirla)')
plt.xlabel('Disminución del Coeficiente de Determinación')
plt.show()
```

La variable más importante es PUNT_INGLES, ya que al ser eliminada, causa la mayor disminución en el R^2 . Esto indica que esa variable aporta información única que las otras no pueden explicar por sí solas.

```python
# 5. Diagrama con el plano de dispersión
#  Las dos variables más explicativas fueron:
var_x = 'PUNT_INGLES'
var_y = 'PUNT_MATEMATICAS'
var_z = 'ESTU_INSE_INDIVIDUAL'

# Crear la figura y el subgráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Tomamos una muestra aleatoria para que el gráfico no sea demasiado pesado
muestra = icfes_sin_nan.sample(n=2000, random_state=42)

# Graficar los puntos
scatter = ax.scatter(muestra[var_x],
                     muestra[var_y],
                     muestra[var_z],
                     c=muestra[var_z], # Color basado en la respuesta
                     cmap='viridis',
                     alpha=0.6)

# Etiquetas de los ejes
ax.set_xlabel(var_x)
ax.set_ylabel(var_y)
ax.set_zlabel('Índice Socioeconómico')
ax.set_title('Plano de Dispersión 3D: Variables más Explicativas')

# Añadir barra de color
fig.colorbar(scatter, ax=ax, label='Nivel Socioeconómico', shrink=0.5)

plt.show()
```

El gráfico muestra una tendencia clara donde el índice socioeconómico (ESTU_INSE_INDIVIDUAL) aumenta de forma proporcional al desempeño en Inglés y Matemáticas.

```python
#6. modelo de regresión lineal múltiple
# Definición de variables finales
X = icfes_sin_nan[['PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS',
                   'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES']]
y = icfes_sin_nan['ESTU_INSE_INDIVIDUAL']

# División 80% entrenamiento, 20% prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)


modelo_lineal = LinearRegression()
modelo_lineal.fit(X_train, y_train)

# Predicciones y métricas
pred_lin = modelo_lineal.predict(X_test)
print(f"R^2 Regresión Lineal: {r2_score(y_test, pred_lin):.4f}")
```

Este modelo explica aproximadamente el 31.98% de la variabilidad del índice socioeconómico. Un $R^2$ superior a 0.30 se considera aceptable, indicando que los puntajes académicos sí tienen una relación lineal significativa con el entorno del estudiante

```python
#7. modelo de red neuronal de perceptrón multicapa
# Configuración según especificaciones del taller [cite: 43, 44]
red = MLPRegressor(
    hidden_layer_sizes=(100, 100, 100, 100, 100, 100),
    max_iter=500,
    activation='relu',
    solver='adam',
    learning_rate='invscaling',
    random_state=42
)

red.fit(X_train, y_train)

# Predicciones y métricas
pred_red = red.predict(X_test)
print(f"R^2 Red Neuronal: {r2_score(y_test, pred_red):.4f}")
```

El modelo complejo explica el 29.76% de la variabilidad. Sorprendentemente, es inferior al modelo lineal por un 2.22%, lo que significa que la red no logró generalizar mejor los patrones de los datos a pesar de su complejidad.

```python

# Definimos los criterios de interés
sobresaliente_lectura = 75
bajo_ingles = 45

# Filtramos el grupo de interés en la base de datos limpia
grupo_especifico = icfes_sin_nan[
    (icfes_sin_nan['PUNT_LECTURA_CRITICA'] >= sobresaliente_lectura) &
    (icfes_sin_nan['PUNT_INGLES'] <= bajo_ingles)
]
# Definimos la lista de variables explicativas que se usaron para entrenar el modelo
variables_expl = ['PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES']

# Realizamos la predicción de su nivel socioeconómico
if not grupo_especifico.empty:
    predicciones_grupo = modelo_lineal.predict(grupo_especifico[variables_expl])
    print(f"Número de estudiantes encontrados: {len(grupo_especifico)}")
    print(f"Promedio de Nivel Socioeconómico Predicho: {predicciones_grupo.mean():.2f}")
else:
    print("No se encontraron estudiantes con ese perfil exacto.")
```
